# Linear regression with gradient descent

```python
slope = random.random()
```

```python
intercept = random.random() 
```

```python
y_ = slope * x + intercept
```

```python
res = y - y_
lsr = sum(res**2)
lsr
```

```python
def lsrf(y , y_):
    res = y - y_
    lsr = sum(res**2)
    return lsr,res
```

```python
def gradient_des(slope,intercept,x,y,lr=0.0001,it=1000):
    m= len(x)
    op=[0,0]
    for i in range(it):
        y_ = slope * x + intercept
#         lsr,res = lsrf(y,y_)
        op[0] = ((-2/m) *sum(x*(y - y_)))                    
        op[1] = ((-2/m) *sum(y - y_))
        print(op)
        slope = slope-lr*op[0]
        intercept = intercept-lr*op[1]
    return slope,intercept
        
```

```python
slope , intercept = gradient_des(slope,intercept,x,y,it=10000)
com.append([slope , intercept])
```

```python
y_ = slope * x + intercept
```

```python
plt.scatter(x,y)
plt.plot(x,y_,color="red")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x1 = x.reshape((len(x),1))
y1 = y.reshape((len(x),1))
X_train, X_test, y_train, y_test = train_test_split(x1, y1, test_size = 0.025)
regr = LinearRegression()

regr.fit(X_train, y_train)
y_ = regr.predict(x1)
```

```python
print(regr.intercept_)
print(regr.coef_)
com.append([regr.coef_[0][0],regr.intercept_[0]])
```

```python
plt.scatter(x,y)
plt.plot(x,y_,color="red")
```

# Linear Regression From Scratch

```python
slope = (len(x)*((x@y.T))-(sum(x)*sum(y)))/((len(x)*(x@x.T))-sum(x)**2)
```

```python
intercept = np.mean(y) - slope*np.mean(x)
```

```python
y_ = slope*x+intercept
```

```python
com.append([slope,intercept])
```

```python
res = y - y_
lsr = sum(res**2)
lsr
```

```python
plt.scatter(x,y)
plt.plot(x,y_,color="red")
```
